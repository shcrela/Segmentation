## 2.3. Sharpening

### 2.3.1. ...using the difference of Gaussians

Here, we try to accentuate the borders by first blurring the image a bit, then substracting the blurred image from the original one which should stress out the edges

```python
# enhancing edges using the diffference of gaussians on both height and phase image
blurred_small = ndimage.gaussian_filter(img_cdb, params['gaussian_blurr_small'])
blurred_large = ndimage.gaussian_filter(img_cdb, params['gaussian_blurr_large'])

blurred_small_ph = ndimage.gaussian_filter(imph_cdb, params['gaussian_blurr_small'])
blurred_large_ph = ndimage.gaussian_filter(imph_cdb, params['gaussian_blurr_large'])
# using the negation (~) to invert the bright and dark regions
edges = ~img_as_uint(blurred_large - blurred_small)
edges_phase = ~img_as_uint(blurred_large_ph-blurred_small_ph)
edges = clahe.apply(edges)
edges_phase = clahe.apply(edges_phase)


```
